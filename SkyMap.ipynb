# Map of Repeating FRB Sources

```python
import pandas as pd
from matplotlib import pyplot as plt
import astropy.coordinates as coord
import astropy.units as u

%matplotlib inline 
sources = pd.read_csv('SourceLocations.csv').set_index('name')
sources['ra_radian'] = coord.SkyCoord(sources['RA'], sources['dec'], unit=(u.hourangle, u.deg), frame='icrs').ra.radian
sources['dec_radian'] = coord.SkyCoord(sources['RA'], sources['dec'], unit=(u.hourangle, u.deg), frame='icrs').dec.radian
plt.figure(figsize=(15, 9))
ax = plt.axes(projection='mollweide')
plt.grid()
plt.scatter(sources['ra_radian'], sources['dec_radian'])
for k, v in sources.iterrows():
    ax.annotate(k, (v['ra_radian'], v['dec_radian']), xytext=(-3,5), textcoords='offset points', weight='bold')
```
